```python
#CLASS METHOD

It is defined ysing the @claamethod decorator.

takes the class itself as the first parameter (often named "cls" by convention.)

can aceess and modify the class level attribuyes.
commonly used for methods that opearate on the class itself , rather than or instances,

NOTE: self pointing to the object adress.
NOTE: Cls will be pointing to the class name addresss.

Syntax:
class class_name:
     @classmethod
     def methodname(cls.parameter1):

#WHY WE NEED CLASS METHOD.

ACCESSING AND MODIFYING THR CLASS-LEVEL ATTRIBUTES.

class methods are having access to the class itself , 
allowing them to work with class-level attributes and perform opearations that affect the entire classs
2. they are particulrly useful when you have need to set or modify the classs variables shared by all instances of the class .
3. class methods do not have access to instance - specific classs(e.g. instance attributes).
    this makes them suitable oparations that dont have depend on the state of a particular object.
             
```

```python
class Students:
    instuite = "qspiders"
    manager = "sri harsha"                         #STATIC STATES
    loc = "dilshukhunagar"
    course = {"java", "python", "testing", "DA"}

    def __init__(self, name, age, roll, selected_course):      # CONSTRUCTOR
        self.std_name = name
        self.std_age = age                         #NON-STATIC STATES
        self.std_roll_num = roll
        self.selected_course = selected_course

    def access_students_data(self):
        print("----- Student Details -----")
        print(f"Name: {self.std_name}")
        print(f"Age: {self.std_age}")
        print(f"Roll Number: {self.std_roll_num}")
        print(f"Selected Course: {self.selected_course}")
        def modify_students_data(self):
            print(f"\n----------MODIFICATION PAGE---------\n")
            change_course = input("select in {java,python,testing}")
            self.selected_course=change_course
            print(f"course changed successfullyin {change_course}")
    @classmethod
    def access_instuite_details(cls):
        print(f"\n ---------{cls.instuite.upper()} DATA-----\n")
        print(f"INSTUITE_NAME : {cls.instuite}\n MANAGER : {cls.manager}\nLOCATION: {cls.loc} \n COURSE :{cls.course}\n")

    @classmethod
    def modify_instuite_details(cls):
        print(f"\n---------- INSTUITE MODIFICATION PAGE---------\n")
        while True:
            print(f"\n SELECT  1 TO CHANGE LOCATION\n SELECT 2 TO CHANGE THE MANAGER\n SELECT 3 TO EXIT\n")
            select = int(input("\n select a option : "))
            
            match select:
                case 1:
                    new_location = input("Enter the new location: ")
                    cls.loc = new_location
                    print("New location updated")

                case 2:
                    new_manager = input("Enter the new manager: ")
                    cls.manager = new_manager
                    print("New manager updated")

                case 3:
                    print("Thanks for updating. You can exit now.")
                    break

                case _:
                    print("You have chosen the wrong number.")
            

# Creating student object
s1 = Students(name="ashok", age=22, roll=1517, selected_course="DA")
s1.access_students_data()
#by using the class name
Students.access_students_data(s1)

#modifying by using the object name

# print("before modifying")
# s1.access_students_data()
# s1.modify_students_data()

print("after modifying")
s1.access_students_data()
#modifying by using the class name

# print("before modifying")
# Students.access_students_data(s1)
# s1.modify_students_data()
# print("after modifying")
# Students.access_students_data(s1)

#class method

# Students.access_instuite_details()
s1.access_instuite_details()


#modification
Students.modify_instuite_details()
Students.access_instuite_details()

s1.access_instuite_details()


```

```python
class Myutility:
    @staticmethod
    def _is_even(n):
        return n%2==0
    @staticmethod
    def _is_odd(n):
        return n%2!=0
    @staticmethod
    def _is_positive(n):
        return n>0
    @staticmethod
    def _is_negiive(n):
        return n<0
    @staticmethod
    def _is_neutral(n):
        return n==0
    @staticmethod
    def validate_age(age):
        if age>=18:
            return age
        else:
            raise exception("not eligible")
s1 = Myutility()
res = s1.is_even(10)
print(res)
print(s1.is_even(10))
print(s1.is_odd(7))
res=s1.validate_age(12)
print(res)


        
        
        
    
```

```python
# MULTI-LEVEL INHERITANCE

class GrandPa:
    g_name="nageswarao"
    def __init__(self,g_property):
        self.g_property=g_property

    def grandpa_details(self):
        print(f"GRANDPA NAME :{self.g_name}\n GRANDPA_PROPERTY :{self.g_property}\n")

class Father(GrandPa):
    f_name="nagerjuna"
    
    def __init__(self,f_property,g_property):
        super().__init__(g_property)
        self.f_property=f_property

    def father_details(self):
        print(f"FATHER NAME :{self.f_name}\n FATHER_PROPERTY :{self.f_property}\n")

class Child(Father):
    c_name="naga chaitanya"
    
    def __init__(self, c_property, f_property, g_property):
        super().__init__(f_property, g_property)
        self.c_property=c_property

    def child_details(self):
        print(f"CHILD NAME :{self.c_name}\n CHILD_PROPERTY :{self.c_property}\n")

c1 = Child(c_property= "10cr", f_property= "100cr" , g_property= "1000cr")

c1.child_details()
c1.father_details()
c1.grandpa_details()


        
        
```

```python
#MULTPLE INHERITANCE
class Father:
    f_name="ram charan"
    
    def __init__(self,f_property):
    
        self.f_property=f_property

    def father_details(self):
        print(f"FATHER NAME :{self.f_name}\n FATHER_PROPERTY :{self.f_property}\n")

class Mother:
    m_name="upasana"
    
    def __init__(self,m_property):
    
        self.f_property=m_property

    def mother_details(self):
        print(f"MOTHER NAME :{self.f_name}\n MOTHER_PROPERTY :{self.f_property}\n")

class Child(Father, Mother):
    m_name="kiln kara"
    
    def __init__(self,c_property, f_property, m_property):
        Father.__init__(self,f_property)
        Mother.__init__(self,m_property)
        
    
        self.c_property = c_property

    def child_details(self):
        print(f"CHILD NAME :{self.f_name}\n CHILD_PROPERTY :{self.f_property}\n")

c1=Child(c_property="1 company",f_property="1000cr", m_property="500cr")

c1.child_details()
c1.father_details()
c1.mother_details()


```

```python

#partial method overloading by using the default parameters.
class Count:
    def add(self,a=0,b=0,c=0):
        print(a+b+c)
c1=Count()
c1.add(a=10)
c1.add(a=10,b=20)
c1.add(a=10,b=20,c=30)


#################################################################################33

#partial method overloading by using the tuple variable len argument(*args)

class Addition:
    def add(self,*args):
        total=0
        for ele in args:
            total +=ele
        print(total)

a1=Addition()
a1.add(10)
a1.add(10,20)
a1.add(10,20,30,30,40)

################################################################################

#partial method overloading by using the dict variable len argument(**kwargs)

class Sample:
    def add(self,**kwargs):
        total=0
        for v in kwargs.values():
            total +=v
        print(total)
s1=Sample()
s1.add(a=10)
s1.add(a=10,b=20)
s1.add(a=10,b=20,c=30,d=30,e=40)

```

```python
#PUBLIC CLASS

class Public:

    a=100
    def __init__(self,b):
        self.b=b

        self.display_public_data()
    def display_public_data(self):

        print(self.a, self.b , sep="|")

p1=Public(200)
print(p1.a)
print(p1.b)
p1.display_public_data()

class SubClass(Public):

    def accessing_public_data(self):

        print(f"CLASS VARIABLE :{self.a}\n"
              f"INSTANCE VARIABLE : {self.b}\n")

        self.display_public_data()

s1=SubClass(200)

s1.accessing_public_data()

```

```python
#PROTECTED
class Protected:
    _a = 100

    def __init__(self, b):
        self._b = b

    def _display_protected_data(self):
        print(self._a, self._b, sep="|")

p1=Protected(200)


class SubClass(Protected):
    def accessing_protected_data(self):
        print(f"class variable : {self._a}\n"
              f"instance variable : {self._b}\n")


s1 = SubClass(200)
s1.accessing_protected_data()
```

```python
#PRIVATE
class Private:
    __a = 100  # Private class variable

    def __init__(self, b):
        self.__b = b  # Private instance variable
        self.__display_private_data()  # Call to private method

    def __display_private_data(self):
        print(self.__a, self.__b, sep="|")


class SubClass(Private):
    def accessing_private_data(self):
        # Accessing private members using name mangling
        print(f"class variable : {self._Private__a}\n"
              f"instance variable : {self._Private__b}\n")
        self._Private__display_private_data()


# Creating object
s1 = SubClass(200)
s1.accessing_private_data()
```
